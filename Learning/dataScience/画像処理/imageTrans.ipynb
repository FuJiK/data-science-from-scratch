```python
# Image1 = misc.face()
# misc.imsave('./gazou/ThinkstockPhotos-100866708-600x400.jpg',Image1)
Image1 = misc.imread('./gazou/parrot.jfif')
# Image1 = sp.misc.imread('./gazou/ThinkstockPhotos-100866708-600x400.jpg')

plt.imshow(Image1)
plt.show()
```

```python
r,g,b = Image1.transpose(2,0,1)
```

```python
np.sum(np.abs(r - Image1[:,:,0])) +  np.sum(np.abs(g - Image1[:,:,1])) + np.sum(np.abs(b - Image1[:,:,2]))
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
#import numpy as np
#import cv2
```

```python
flg, axes = plt.subplots(ncols=4, figsize=(20,8))
axes[0].set_title('natural color')
axes[0].imshow(Image1)
axes[0].get_xaxis().set_visible(False)
axes[0].get_yaxis().set_visible(False)

axes[1].set_title('R loss')
axes[1].imshow(r, cmap=cm.Greys_r, vmin=0, vmax=255)
axes[1].get_xaxis().set_visible(False)
axes[1].get_yaxis().set_visible(False)

axes[2].set_title('G loss')
axes[2].imshow(g, cmap=cm.Greys_r, vmin=0, vmax=255)
axes[2].get_xaxis().set_visible(False)
axes[2].get_yaxis().set_visible(False)

axes[3].set_title('B loss')
axes[3].imshow(b, cmap=cm.Greys_r, vmin=0, vmax=255)
axes[3].get_xaxis().set_visible(False)
axes[3].get_yaxis().set_visible(False)
```
